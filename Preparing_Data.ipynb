```python
import os
import requests
import tiktoken
import numpy as np
from data.data_path_retrieve import data_path_retrieve

# download the tiny shakespeare dataset
input_file_path = data_path_retrieve() + "/input.txt"
if not os.path.exists(input_file_path):
    data_url = 'https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt'
    with open(input_file_path, 'w', encoding='utf-8') as f:
        f.write(requests.get(data_url).text)
```

```python
# 읽어오기
## 보통은 다운로드 받은 것을 사용할테니까, 불러오는 환경에 맞춰서 진행하도록 하자
    ### 이렇게 하면 다운로드 받은 데이터도 이상이 없는지 검토하는 과정도 될 수 있다..
with open(input_file_path, 'r', encoding='utf-8') as f:
    data = f.read()
```

```python
n = len(data)
train_val_split_index = int(n*0.9)
train_set = data[:train_val_split_index]
val_set = data[train_val_split_index:]
```

```python
n == (len(train_set) + len(val_set)), data == (train_set + val_set)
```

```python
enc = tiktoken.get_encoding("gpt2")
```

```python
train_ids = enc.encode_ordinary(train_set)
val_ids = enc.encode_ordinary(val_set)
```

```python
print(f"train has {len(train_ids):,} tokens")
print(f"val has {len(val_ids):,} tokens")
```

```python
train_ids = np.array(train_ids, dtype=np.uint16)
val_ids = np.array(val_ids, dtype=np.uint16)
```

```python
train_ids.tofile(os.path.join(data_path_retrieve(), 'train.bin'))
val_ids.tofile(os.path.join(data_path_retrieve(), 'val.bin'))
```
